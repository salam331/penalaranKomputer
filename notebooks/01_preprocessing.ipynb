```python
from pathlib import Path
from pdfminer.high_level import extract_text

pdf_folder = Path("../data/pdf")
raw_folder = Path("../data/raw")
raw_folder.mkdir(parents=True, exist_ok=True)

print("PDF folder:", pdf_folder.resolve())
print("Exists:", pdf_folder.exists())
print("PDF ditemukan:", len(list(pdf_folder.glob("*.pdf"))))

for pdf_file in pdf_folder.glob("*.pdf"):
    try:
        text = extract_text(pdf_file)

        txt_file = raw_folder / (pdf_file.stem + ".txt")

        with open(txt_file, "w", encoding="utf-8") as f:
            f.write(text)

        print(f"Berhasil: {pdf_file.name}")

    except Exception as e:
        print(f"Gagal: {pdf_file.name}")
        print(e)
```

```python
import re
```

```python
def clean_text(text):

    text = re.sub(
        r'halaman\s+\d+',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'disclaimer',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'mahkamah agung republik indonesia',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'\s+',
        ' ',
        text
    )

    text = text.lower()

    return text.strip()
```

```python
for txt_file in raw_folder.glob("*.txt"):

    with open(
        txt_file,
        "r",
        encoding="utf-8"
    ) as f:

        text = f.read()

    cleaned_text = clean_text(text)

    with open(
        txt_file,
        "w",
        encoding="utf-8"
    ) as f:

        f.write(cleaned_text)

print("Cleaning selesai")
```
