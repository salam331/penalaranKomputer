```python
from pdfminer.high_level import extract_text
from pathlib import Path
import re
```

```python
pdf_folder = Path("data/pdf")
raw_folder = Path("data/raw")

print("Folder PDF :", pdf_folder)
print("Folder RAW :", raw_folder)
```

```python
raw_folder.mkdir(
    parents=True,
    exist_ok=True
)

print("Folder raw siap digunakan")
```

```python
pdf_files = list(pdf_folder.glob("*.pdf"))

print("Jumlah PDF:", len(pdf_files))

for file in pdf_files[:5]:
    print(file.name)
```

```python
for pdf_file in pdf_files:

    try:

        text = extract_text(pdf_file)

        txt_name = pdf_file.stem + ".txt"

        txt_path = raw_folder / txt_name

        with open(
            txt_path,
            "w",
            encoding="utf-8"
        ) as f:

            f.write(text)

        print(
            f"✓ Berhasil: {pdf_file.name}"
        )

    except Exception as e:

        print(
            f"✗ Gagal: {pdf_file.name}"
        )

        print(e)
```

```python
from pathlib import Path

txt_files = list(Path("../data/raw").glob("*.txt"))

print(len(txt_files))
print(txt_files[:5])
```

```python
def clean_text(text):

    text = re.sub(
        r'halaman\s+\d+',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'disclaimer',
        ' ',
        text,
        flags=re.IGNORECASE
    )

    text = re.sub(
        r'\s+',
        ' ',
        text
    )

    text = text.lower()

    return text.strip()
```

```python
for txt_file in txt_files:

    with open(
        txt_file,
        "r",
        encoding="utf-8"
    ) as f:

        text = f.read()

    cleaned_text = clean_text(text)

    with open(
        txt_file,
        "w",
        encoding="utf-8"
    ) as f:

        f.write(cleaned_text)

print("Cleaning selesai")
```
