```python
import pandas as pd
import numpy as np
from pathlib import Path
import re
```

```python
df = pd.read_csv("../data/processed/cases.csv")

print(df.shape)
df.head()
```

```python
print(df.columns)
```

```python
df["text_full"] = df["text_full"].fillna("")
df["nomor_perkara"] = df["nomor_perkara"].fillna("unknown")
df["jenis_perkara"] = df["jenis_perkara"].fillna("unknown")
```

```python
df["panjang_teks"] = df["text_full"].apply(lambda x: len(str(x).split()))
```

```python
def simple_summary(text, max_words=80):
    words = str(text).split()
    return " ".join(words[:max_words])

df["ringkasan_fakta"] = df["text_full"].apply(simple_summary)
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"[^a-z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["text_clean"] = df["text_full"].apply(clean_text)
```

```python
df_final = df[[
    "case_id",
    "nomor_perkara",
    "jenis_perkara",
    "ringkasan_fakta",
    "panjang_teks",
    "text_full"
]]

df_final.head()
```

```python
Path("../data/processed").mkdir(parents=True, exist_ok=True)

df_final.to_csv(
    "../data/processed/cases.csv",
    index=False,
    encoding="utf-8-sig"
)

print("Dataset Case Representation tersimpan!")
```
