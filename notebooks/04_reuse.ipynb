```python
import re

def extract_amar(text):

    text = str(text).lower()

    patterns = [
        r"amar\s+putusan(.*?)(menyatakan|memutuskan|demikianlah)",
        r"memutuskan(.*?)(demikian|ditetapkan)",
        r"mengadili(.*?)(demikian|ditetapkan)"
    ]

    for pattern in patterns:

        match = re.search(pattern, text, re.DOTALL)

        if match:
            return match.group(0)[:2000]

    return "AMAR TIDAK DITEMUKAN"
```

```python
df["amar_putusan"] = df["text_full"].apply(extract_amar)

df[["case_id", "amar_putusan"]].head()
```

```python
case_solutions = {}

for _, row in df.iterrows():
    case_solutions[row["case_id"]] = row["amar_putusan"]

print("Total cases:", len(case_solutions))
```

```python
def get_top_solutions(query, k=5):

    top_cases = retrieve(query, k=k)

    solutions = []

    for cid in top_cases["case_id"]:
        solutions.append(case_solutions[cid])

    return solutions
```

```python
query = "pengujian undang undang kehutanan"

get_top_solutions(query)
```

```python
def predict_majority(query, k=5):

    solutions = get_top_solutions(query, k)

    vote = Counter(solutions)

    return vote.most_common(1)[0][0]
```

```python
predict_majority(
    "pengujian undang undang kehutanan"
)
```

```python
def predict_weighted(query, k=5):

    top_cases = retrieve(query, k=k)

    scores = {}

    for _, row in top_cases.iterrows():

        cid = row["case_id"]
        sim = row["similarity"]

        sol = case_solutions[cid]

        scores[sol] = scores.get(sol, 0) + sim

    return max(scores, key=scores.get)
```

```python
predict_weighted(
    "pengujian undang undang kehutanan"
)
```

```python
queries = [
    "pengujian undang undang kehutanan",
    "perselisihan hasil pemilu presiden",
    "larangan pemakaian tanah tanpa izin"
]
```

```python
results = []

for q in queries:

    pred = predict_weighted(q)

    top5 = retrieve(q, k=5)

    results.append({
        "query": q,
        "predicted_solution": pred,
        "top_5_case_ids": ",".join(top5["case_id"])
    })

pd.DataFrame(results)
```

```python
output_path = "../data/results/predictions.csv"

import os
os.makedirs("../data/results", exist_ok=True)

pd.DataFrame(results).to_csv(
    output_path,
    index=False,
    encoding="utf-8-sig"
)

print("Saved:", output_path)
```
